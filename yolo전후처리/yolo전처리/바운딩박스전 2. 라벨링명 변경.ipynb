```python
import os.path
import pandas as pd
os.chdir(path)
def read_data(filename, encoding='cp949'):
# 읽기 함수 정의
    with open(filename, 'r', encoding=encoding) as f:
        data = [line.split('\t') for line in f.read().splitlines()]
        d1 = ''
        for i in data:
          i[0]=i[0].replace('None','Pig')
        for line in data[1:]:
          line[0]=line[0].replace('Pig','\n' + 'Pig')
        for line in data:
            d1 += line[0]
        data = d1
    print(data)
    return data
def write_data(data, filename, encoding='cp949'):
    with open(filename, 'w', encoding=encoding) as f:
        f.write(data)
```

```python
for cnt, line in enumerate(file_list_py):
  write_data(str(read_data(line)),line)
```
