## yolo txt파일 coco json으로 변환

```python
# 필요 라이브러리 설치 및 불러오기
import logging
logging.getLogger().setLevel(logging.CRITICAL)
!pip install pylabel > /dev/null
```

```python
# 필요 라이브러리 불러오기
from pylabel import importer
```

```python
# 실제 변환 코드
path_to_annotations = "/content/drive/MyDrive/yolov7/runs/detect/exp3/labels" # txt파일이 있는 경로 설정
path_to_images = "/content/drive/MyDrive/yolov7/data/test/images/" # 이미지 파일이 있는 경로 설정

yoloclasses = ['Pig', '0'] # 라벨명 설정

dataset = importer.ImportYoloV5(path=path_to_annotations, path_to_images=path_to_images, cat_names=yoloclasses,
    img_ext="jpg", name="coco128") # 필요사항 설정(name에 파일명 설정)
```

```python
# 저장(경로는 path경로에 저장됨)
dataset.export.ExportToCoco(cat_id_index=1)
```
